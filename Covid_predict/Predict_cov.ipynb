```python

countrys= data["Country/Region"]
def get_index_country(name):
    for i in range(len(countrys)):
        if(countrys[i]=="Cuba"):
           return i
    return -1
def convert_format_time(date):
    month,day,year = date.split("/")
    if(len(month)<2):
        month="0"+month
    if(len(day)<2):
        day="0"+day
    return "20"+year+"-"+month+"-"+day
```

```python

def get_number_case_of_covid(country_name):
    data_ncoc=None
    index_country=get_index_country(country_name)
    if(index_country!=-1):
        data_ncoc={}
        dates=data.iloc[index_country][4:].keys()
        values=data.iloc[index_country][4:]
        for i in range(len(dates)):
            data_ncoc[convert_format_time(dates[i])]=values[i]
    return data_ncoc
        
```

```python
# input x1(cases of covid19),x2(corona).....x13(covid19 cases) 
# output y(number case of cov)
X_total=[]
Y_total=[]
for name_contry in data_total.keys():
    months=list(data_total[name_contry].keys())
    for i in range(len(months)-1):
        month = months[i]
        next_month =months[i+1]
        x=[]
        for kw in keywords:
            if(kw not in data_total[name_contry][month]):
                x.append(0)
            else:
                x.append(data_total[name_contry][month][kw])
        y=data_total[name_contry][next_month]["value"]
        X_total.append(x)
        Y_total.append(y)
        
```
